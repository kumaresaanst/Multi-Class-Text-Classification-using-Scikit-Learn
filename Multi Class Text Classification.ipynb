```python
import pandas as pd
import numpy as np
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score,roc_auc_score,confusion_matrix
```

```python
train = pd.read_csv('C:/Users/HP/Downloads/Selection_sklearn/Machine_Learning_with_Text/pycon-2016-tutorial-master/data/yelp.csv')
```

```python
X = train['text']
y = train['stars']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1)
X_train.shape
```

```python
y_train.value_counts()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(min_df=5,stop_words='english',ngram_range=(1,2),max_df=0.9)
```

```python
X_train_dtm = vect.fit_transform(X_train)
X_test_dtm = vect.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB(alpha=1)
```

```python
nb.fit(X_train_dtm,y_train)
```

```python
pre = nb.predict(X_test_dtm)
```

```python
accuracy_score(y_test,pre)
```

```python
X_train_tokens = vect.get_feature_names()
```

```python
confusion_matrix(y_test,pre)
```

```python
X_test[pre<y_test].head()
```

```python
nb.feature_count_
```

```python
onestarcount = nb.feature_count_[0,:]
twostarcount = nb.feature_count_[1,:]
threestarcount=nb.feature_count_[2,:]
fourstarcount=nb.feature_count_[3,:]
fivestarcount = nb.feature_count_[4,:]
```

```python
onestarcount
```

```python
reviews = pd.DataFrame({'Token':X_train_tokens,'1starcount':onestarcount,'2starcount':twostarcount,'3starcount':threestarcount,'4starcount':fourstarcount,'5starcount':fivestarcount}).set_index('Token')
```

```python
reviews['1starcount']=reviews['1starcount']/nb.class_count_[0]
reviews['2starcount']=reviews['2starcount']/nb.class_count_[1]
reviews['3starcount']=reviews['3starcount']/nb.class_count_[2]
reviews['4starcount']=reviews['4starcount']/nb.class_count_[3]
reviews['5starcount']=reviews['5starcount']/nb.class_count_[4]
```

```python
reviews['1starcount'] = reviews['1starcount']+1
reviews['2starcount'] = reviews['2starcount']+1
reviews['3starcount'] = reviews['3starcount']+1
reviews['4starcount'] = reviews['4starcount']+1
reviews['5starcount'] = reviews['5starcount']+1
```

```python
reviews['1starratio']=reviews['1starcount']/(reviews['2starcount']+reviews['3starcount']+reviews['4starcount']+reviews['5starcount'])
reviews['2starratio']=reviews['2starcount']/(reviews['1starcount']+reviews['3starcount']+reviews['4starcount']+reviews['5starcount'])
reviews['3starratio']=reviews['3starcount']/(reviews['1starcount']+reviews['2starcount']+reviews['4starcount']+reviews['5starcount'])
reviews['4starratio']=reviews['4starcount']/(reviews['1starcount']+reviews['2starcount']+reviews['3starcount']+reviews['5starcount'])
reviews['5starratio']=reviews['5starcount']/(reviews['1starcount']+reviews['2starcount']+reviews['3starcount']+reviews['4starcount'])
```

```python
onestar = reviews[(reviews['1starratio']>reviews['2starratio']) & (reviews['1starratio']>reviews['3starratio']) & (reviews['1starratio']>reviews['4starratio']) & (reviews['1starratio']>reviews['5starratio'])]
twostar = reviews[(reviews['2starratio']>reviews['1starratio']) & (reviews['2starratio']>reviews['3starratio']) & (reviews['2starratio']>reviews['4starratio']) & (reviews['2starratio']>reviews['5starratio'])]
threestar = reviews[(reviews['3starratio']>reviews['1starratio']) & (reviews['3starratio']>reviews['2starratio']) & (reviews['3starratio']>reviews['4starratio']) & (reviews['3starratio']>reviews['5starratio'])]
fourstar = reviews[(reviews['4starratio']>reviews['1starratio']) & (reviews['4starratio']>reviews['2starratio']) & (reviews['4starratio']>reviews['3starratio']) & (reviews['4starratio']>reviews['5starratio'])]
fivestar = reviews[(reviews['5starratio']>reviews['1starratio']) & (reviews['5starratio']>reviews['2starratio']) & (reviews['5starratio']>reviews['3starratio']) & (reviews['5starratio']>reviews['4starratio'])]
```
